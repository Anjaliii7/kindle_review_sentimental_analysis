```python
# data preprocessing and cleaning
# positive review is 1 and negative review is 0
data['rating']=data['rating'].apply(lambda x: 0 if x<3 else 1)
```

```python
data.head()
```

```python
data['rating'].value_counts()
```

```python
# lower all the cases
data['reviewText']=data['reviewText'].str.lower()
```

```python
data.head()
```

```python
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from bs4 import BeautifulSoup
```

```python
# removing all special characters
data['reviewText']=data['reviewText'].apply(lambda x:re.sub('[^a-z A-Z 0-9]+',' ',x))
```

```python
# removing all stopwords
data['reviewText']=data['reviewText'].apply(lambda x: " ".join([y for y in x.split() if y.lower() not in stopwords.words('english')]))
```

```python
# remove all url's
data['reviewText']=data['reviewText'].apply(lambda x: re.sub(r"http\S+", "", x))
```

```python
# remove html tags
data['reviewText']=data['reviewText'].apply(lambda x: BeautifulSoup(x,'lxml').get_text())
```

```python
# remove extra spaces
data['reviewText']=data['reviewText'].apply(lambda x: " ".join(x.split()))
```

```python
data.head()
```

```python
# apply lemmatization
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
Lemmatizer=WordNetLemmatizer()
data['reviewText']=data['reviewText'].apply(lambda x: " ".join([Lemmatizer.lemmatize(y) for y in x.split()]))
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data['reviewText'],data['rating'],test_size=0.2,random_state=0)
```

```python
# bag of words
from sklearn.feature_extraction.text import CountVectorizer
bow=CountVectorizer()
x_train_bow=bow.fit_transform(x_train).toarray()
x_test_bow=bow.transform(x_test).toarray()
```

```python
# TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
x_train_tfidf=tfidf.fit_transform(x_train).toarray()
x_test_tfidf=tfidf.transform(x_test).toarray()
```

```python
x_train_bow
```

```python
# naive bayes
from sklearn.naive_bayes import GaussianNB
nb_model_bow=GaussianNB().fit(x_train_bow,y_train)
nb_model_tfidf=GaussianNB().fit(x_train_tfidf,y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
y_pred_bow=nb_model_bow.predict(x_test_bow)
```

```python
y_pred_tfidf=nb_model_tfidf.predict(x_test_tfidf)
```

```python
print("BOW Accuracy :",accuracy_score(y_test,y_pred_bow))
```

```python
confusion_matrix(y_test,y_pred_bow)
```

```python
print("TF-IDF Accuracy :",accuracy_score(y_test,y_pred_tfidf))
```

```python
confusion_matrix(y_test,y_pred_tfidf)
```

```python
def average_word2vec(tokens_list, vector, generate_missing=False, k=100):
    if len(tokens_list)<1:
        return np.zeros(k)
    if generate_missing:
        vectorized = [vector[word] if word in vector else np.random.rand(k) for word in tokens_list]
    else:
        vectorized = [vector[word] if word in vector else np.zeros(k) for word in tokens_list]
    length = len(vectorized)
    sum_vectors = sum(vec for vec in vectorized)
    return np.array(sum_vectors) / length
```
